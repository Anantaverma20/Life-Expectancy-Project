# 8) Simple Linear Regression
---

```r
# Load the required library for plotting
library(ggplot2)

# Model 1: GDP as a Predictor of Life Expectancy
model1 <- lm(Life.expectancy ~ GDP, data = Data.life.expectancy)
summary(model1)

# Create scatterplot with regression line
ggplot(Data.life.expectancy, aes(x = GDP, y = Life.expectancy)) +
  geom_point() +
  geom_smooth(method = "lm", formula = y ~ x, se = FALSE, color = "blue") +
  labs(x = "GDP", y = "Life Expectancy") +
  ggtitle("Linear Regression: GDP vs. Life Expectancy")
```

### Model 1: GDP as a Predictor of Life Expectancy

Interpretation: There is a positive relationship between a country's Gross Domestic Product (GDP) and life expectancy. As GDP increases, life expectancy tends to increase as well.

```r
# Model 2: Schooling as a Predictor of Life Expectancy
model2 <- lm(Life.expectancy ~ Schooling, data = Data.life.expectancy)
summary(model2)

# Create scatterplot with regression line
ggplot(Data.life.expectancy, aes(x = Schooling, y = Life.expectancy)) +
  geom_point() +
  geom_smooth(method = "lm", formula = y ~ x, se = FALSE, color = "red") +
  labs(x = "Schooling", y = "Life Expectancy") +
  ggtitle("Linear Regression: Schooling vs. Life Expectancy")
```

### Model 2: Schooling as a Predictor of Life Expectancy

Interpretation: There is a positive relationship between the average years of schooling in a country and life expectancy. As the level of education (schooling) increases, life expectancy tends to rise.

```r
# Model 3: Alcohol Consumption as a Predictor of Life Expectancy
model3 <- lm(Life.expectancy ~ Alcohol, data = Data.life.expectancy)
summary(model3)

# Create scatterplot with regression line
ggplot(Data.life.expectancy, aes(x = Alcohol, y = Life.expectancy)) +
  geom_point() +
  geom_smooth(method = "lm", formula = y ~ x, se = FALSE, color = "green") +
  labs(x = "Alcohol Consumption", y = "Life Expectancy") +
  ggtitle("Linear Regression: Alcohol vs. Life Expectancy")
```

### Model 3: Alcohol Consumption as a Predictor of Life Expectancy

Interpretation: This model which opposite that shows negative relationship between alcohol consumption and life expectancy. Countries with higher alcohol consumption tend to have lower life expectancies.

```r
# Model 4: HIV/AIDS Prevalence as a Predictor of Life Expectancy
model4 <- lm(Life.expectancy ~ HIV.AIDS, data = Data.life.expectancy)
summary(model4)

# Create scatterplot with regression line
ggplot(Data.life.expectancy, aes(x = HIV.AIDS, y = Life.expectancy)) +
  geom_point() +
  geom_smooth(method = "lm", formula = y ~ x, se = FALSE, color = "purple") +
  labs(x = "HIV/AIDS Prevalence", y = "Life Expectancy") +
  ggtitle("Linear Regression: HIV/AIDS vs. Life Expectancy")
```

### Model 4: HIV/AIDS Prevalence as a Predictor of Life Expectancy

Interpretation: There is a negative relationship between HIV/AIDS prevalence and life expectancy. Higher HIV/AIDS prevalence is associated with lower life expectancy.

```r
# Model 5: Income Composition of Resources as a Predictor of Life Expectancy
model5 <- lm(Life.expectancy ~ Income.composition.of.resources, data = Data.life.expectancy)
summary(model5)

# Create scatterplot with regression line
ggplot(Data.life.expectancy, aes(x = Income.composition.of.resources, y = Life.expectancy)) +
  geom_point() +
  geom_smooth(method = "lm", formula = y ~ x, se = FALSE, color = "orange") +
  labs(x = "Income Composition of Resources", y = "Life Expectancy") +
  ggtitle("Linear Regression: Income Composition vs. Life Expectancy")
```

### Model 5: Income Composition of Resources as a Predictor of Life Expectancy

Interpretation: There is a positive relationship between a country's income composition of resources and life expectancy. As the income rises, life expectancy tends to increase.

# 9) Multiple Linear Regression

---




```r
MLR1 <- lm(Life.expectancy ~ Alcohol+ GDP + Income.composition.of.resources + Adult.Mortality + Schooling+ HIV.AIDS + BMI, data=Data.life.expectancy)
summary(MLR1)
```

```r
cat("Predicted Life Expectancy = " , coef(MLR1)[1],
      "+", coef(MLR1)[2], "* Alcohol)"
     , "+", coef(MLR1)[3], "* (GDP)"
     , "+", coef(MLR1)[4], "* (Income.composition.of.resources)"
     , "+", coef(MLR1)[5], "* (Adult.Mortality)"
     , "+", coef(MLR1)[6], "* (Schooling)"
     , "+", coef(MLR1)[7], "* (HIV.AIDS)"
     , "+", coef(MLR1)[8], "* (BMI)")
```

```r
MLR1.prediction <- predict(MLR1,newdata=data.frame(BMI = 17.7, Alcohol = 0.8, GDP = 673.8227, Income.composition.of.resources = 0.384, Schooling = 7.2, HIV.AIDS = 0.8, Adult.Mortality = 271 ))
MLR1.prediction
```

In MLR2 we will replace the GDP with Mortality.to.Life.Expectancy.Ratio feature and see the changes

```r
MLR2 <- lm(Life.expectancy ~ Alcohol+ Mortality.to.Life.Expectancy.Ratio + Income.composition.of.resources + Adult.Mortality + Schooling+ HIV.AIDS + BMI, data=Data.life.expectancy)
summary(MLR2)
```

The above prediction shows that our MLR model works perfectly for the given parameters.

```r
MLR2.prediction <- predict(MLR2,newdata=data.frame(BMI = 17.7, Alcohol = 0.8, Mortality.to.Life.Expectancy.Ratio = 4.745 , Income.composition.of.resources = 0.384, Schooling = 7.2, HIV.AIDS = 0.8, Adult.Mortality = 271 ))
MLR2.prediction
```

Here in MLR3 we removed BMI and added Vaccination.Coverage and also GDP

```r
MLR3 <- lm(Life.expectancy ~ Alcohol+ Mortality.to.Life.Expectancy.Ratio + Income.composition.of.resources + Adult.Mortality + Schooling+ HIV.AIDS + GDP + Vaccination.Coverage, data=Data.life.expectancy)
summary(MLR3)
```

```r
MLR3.prediction <- predict(MLR3,newdata=data.frame(GDP = 673, Vaccination.Coverage =67,Alcohol = 0.8, Mortality.to.Life.Expectancy.Ratio = 4.745 , Income.composition.of.resources = 0.384, Schooling = 7.2, HIV.AIDS = 0.8, Adult.Mortality = 271 ))
MLR3.prediction
```

In MLR4 infant deaths and hepatitis b were added

```r
MLR4 <- lm(Life.expectancy ~  infant.deaths + Hepatitis.B + Mortality.to.Life.Expectancy.Ratio + Income.composition.of.resources + Adult.Mortality + Schooling+ HIV.AIDS + GDP + Vaccination.Coverage, data=Data.life.expectancy)
summary(MLR4)
```

```r
MLR4.prediction <- predict(MLR4,  newdata=data.frame(infant.deaths = 64, Hepatitis.B = 67, GDP = 673, Vaccination.Coverage =67,Alcohol = 0.8, Mortality.to.Life.Expectancy.Ratio = 4.745 , Income.composition.of.resources = 0.384, Schooling = 7.2, HIV.AIDS = 0.8, Adult.Mortality = 271 ))
MLR4.prediction
```

MLR 5 includes total expenditure and diphtheria columns as compared to MLR4

```r
MLR5.prediction <- predict(MLR5,  newdata=data.frame(Total.expenditure =  8.16 , Diphtheria = 67 , infant.deaths = 64, Hepatitis.B = 67, GDP = 673, Vaccination.Coverage =67,Alcohol = 0.8, Mortality.to.Life.Expectancy.Ratio = 4.745 , Income.composition.of.resources = 0.384, Schooling = 7.2, HIV.AIDS = 0.8, Adult.Mortality = 271 ))
MLR5.prediction
```

We investigated the link between healthcare spending and life expectancy. While the link is complicated, nations with higher per capita healthcare spending have longer life expectancies. This research highlights the importance of investing in healthcare infrastructure.



5. Alcohol Consumption
Another element we looked into was alcohol consumption. Notably, there is a link between alcohol intake and life expectancy. Reducing excessive alcohol intake may be beneficial to public health.




6. Adult Mortality
 Higher adult mortality rates were associated with lower life expectancies. Identifying and addressing factors contributing to adult mortality should be a priority.

7. Income Composition
Income composition of resources was found to have a significant impact on life expectancy. Countries with a higher income composition generally enjoy longer life expectancies, highlighting the importance of economic development.
